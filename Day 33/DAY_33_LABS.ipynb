```python
import pandas as pd

# Input DataFrame
df = pd.DataFrame({
    "Math": [12, 4, 7, None, 2],
    "English": [4, 3, 57, 3, None],
    "Hindi": [20, 16, None, 3, 8],
    "Science": [14, 3, None, None, 6]
})

# Interpolate missing values
df_interpolated = df.interpolate()

# Display the original and interpolated DataFrames
print("Original DataFrame:")
print(df)
print("\nDataFrame after Interpolation:")
print(df_interpolated)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Input data
data = {
    'Subject': ['Math', 'Physics', 'Chemistry', 'Biology'],
    'Student_A': [80, 85, np.nan, 70],
    'Student_B': [90, np.nan, 75, 85],
    'Student_C': [80, 88, np.nan, 78]
}

# Create DataFrame
df = pd.DataFrame(data)

# Interpolate missing values
df_interpolated = df.interpolate()

# Display the original and interpolated DataFrames
print("Original DataFrame:")
print(df)
print("\nDataFrame after Interpolation:")
print(df_interpolated)

# Calculate subject-wise highest score
subject_highest_score = df_interpolated.set_index('Subject').max(axis=1)

# Generate bar chart
plt.figure(figsize=(10, 6))
subject_highest_score.plot(kind='bar', color='skyblue')
plt.title('Subject-wise Highest Score')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.show()
```
